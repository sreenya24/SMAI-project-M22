```python
## SIFT feature extraction
# calculating sift for each block
img = cv.imread('red.jpg')
sift = cv.SIFT_create()
gray = cv.cvtColor(img,cv.COLOR_BGR2GRAY)
kp = []
des = []
k = 128
for i in range(img.shape[0]//k):
    for j in range(img.shape[1]//k):
        kp_block = sift.detect(gray[i*k:(i+1)*k,j*k:(j+1)*k], None)
        # extract only the strongest keypoints in each block
        kp_block = sorted(kp_block, key=lambda x: -x.response)[:1]
        # if there is no keypoint in the block, take the centre point as the keypoint
        if len(kp_block) == 0:
         kp_block = [cv.KeyPoint(i*k+k/2,j*k+k/2,k)]
        kp_block,des_block = sift.compute(gray[i*k:(i+1)*k,j*k:(j+1)*k],kp_block)
        kp.append(kp_block)
        des.append(des_block)
```

```python
import numpy as np
import pandas as pd
from scipy import stats as st
from sklearn.model_selection import train_test_split
```

```python
#Load data
iris = pd.read_csv('Iris.csv')
#data cleaning
iris.drop(columns="Id",inplace=True)
```

```python
#features and labels
X=iris.iloc[:,0:4].values
y=iris.iloc[:,4].values
```

```python
# kmeans from scratch 

def distance(x,y):
    return np.sqrt(np.sum((x-y)**2))

def mean(x):
    return np.sum(x,axis=0)/len(x)

def k_means_clustering(X,k):
    centers = X[np.random.choice(X.shape[0],k,replace=False),:]

    labels = np.zeros(X.shape[0])
    
    dist = np.zeros((X.shape[0],k))
 
    n_iter = 0
    
    n_changes = 0

    prev_labels = np.zeros(X.shape[0])
    while True:
        for i in range(k):
            dist[:,i] = np.apply_along_axis(distance,1,X,centers[i,:])
        labels = np.argmin(dist,axis=1)
       
        for i in range(k):
            centers[i,:] = mean(X[labels==i,:])
  
        n_changes = np.sum(labels!=prev_labels)
    
        prev_labels = labels
       
        n_iter += 1
  
        if n_changes==0:
            break
    return labels,centers,n_iter
```

```python
# call the function to cluster the data
labels,centers,n_iter = k_means_clustering(X,3)

# print the results
print('Number of iterations: ',n_iter)
print('Centers: ',centers)
print('Labels: ',labels)

```
